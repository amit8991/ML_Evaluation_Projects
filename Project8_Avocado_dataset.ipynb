```python
#Separate the dataset into input and output dataset
df_x=df_avocado.drop(columns=["AveragePrice"])
df_y=df_avocado[["AveragePrice"]]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df_x = sc.fit_transform(df_x)
df_x
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maxr2_score(regr,df_x,df_y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(df_x, df_y,random_state = r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
    return final_r_state
```

```python
from sklearn.model_selection import cross_val_score
def model_evaluation(model,df_x,df_y):
    cvs=cross_val_score(model,df_x,df_y,cv=5,scoring="r2")
    print("Mean r2 score : ",cvs.mean())
    print("standard deviation in r2 score for regressor: ",cvs.std())
    print(cvs)
```

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
r_state = maxr2_score(lg,df_x,df_y)
```

```python
model_evaluation(lg,df_x,df_y)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
r_state = maxr2_score(knn,df_x,df_y)
```

```python
model_evaluation(knn,df_x,df_y)
```

```python
from sklearn.model_selection import GridSearchCV
neighbors={"n_neighbors":range(1,25)}
knr=KNeighborsRegressor()
clf = GridSearchCV(knr, neighbors, cv=10)
clf.fit(df_x,df_y)
clf.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=16)
r_state=maxr2_score(knr,df_x,df_y)
```

```python
model_evaluation(knr,df_x,df_y)
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada_knn=AdaBoostRegressor()
ada_knn=AdaBoostRegressor(base_estimator=knn,learning_rate=0.1,n_estimators=10)
```

```python
r_state=maxr2_score(ada_knn,df_x,df_y)
```

```python
#Selecting RandomForestRegressor as final model because of highest r2 score
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y,random_state =90,test_size=0.20)
rf_final=RandomForestRegressor()
rf_final.fit(x_train,y_train)
y_pred=rf_final.predict(x_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print("RMSE is: ",np.sqrt(mean_squared_error(y_test,y_pred)))
print("r2_score is: ",r2_score(y_test,y_pred))
```
