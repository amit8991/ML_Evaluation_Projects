```python
#Check for presence of outliers
from scipy.stats import zscore
z_scr=zscore(df_claims)
print(df_claims.shape)
df_claims_final=df_claims.loc[(abs(z_scr)<3).all(axis=1)]
print(df_claims_final.shape)
```

```python
df_claims_final.skew()
```

```python
#Removing skewness
import numpy as np
for index in df_claims_final.skew().index:
    if df_claims_final.skew().loc[index]>0.5:
        df_claims_final[index]=np.sqrt(df_claims_final[index])
```

```python
df_claims_final.skew()
```

```python
#Seperate the input and output column
df_x=df_claims_final.drop(columns=["Total Claim Amount"])
df_y=df_claims_final[["Total Claim Amount"]]
```

```python
#Standard Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(df_x)
df_x=sc.transform(df_x)
```

```python
df_x
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maxr2_score(regr,df_x,df_y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(df_x, df_y,random_state = r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
    return final_r_state
```

```python
from sklearn.model_selection import cross_val_score
def model_evaluation(model,df_x,df_y):
    cvs=cross_val_score(model,df_x,df_y,cv=5,scoring="r2")
    print("Mean r2 score : ",cvs.mean())
    print("standard deviation in r2 score for regressor: ",cvs.std())
    print(cvs)
```

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
r_state = maxr2_score(lg,df_x,df_y)
```

```python
model_evaluation(lg,df_x,df_y)
```

```python
#AdaBoost regressor
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostRegressor
ada_reg=AdaBoostRegressor(base_estimator =lg)
parameters={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500]}
ada_1 = GridSearchCV(ada_reg,parameters,cv=5)
ada_1.fit(df_x,df_y)
ada_1.best_params_
```
