```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(df_x)
df_x=sc.transform(df_x)
df_x
```

```python
from sklearn.model_selection import train_test_split,cross_val_score,cross_val_predict
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_curve,auc
from sklearn.metrics import f1_score
def maxf1_score(clf,df_x,df_y):
    maxf=0
    rs=0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(df_x, df_y,random_state = r_state,test_size=0.20,stratify=df_y)
        clf.fit(x_train,y_train)
        y_pred=clf.predict(x_test)
        tmp=accuracy_score(y_test,y_pred)
        if tmp>maxf:
            maxf=tmp
            rs=r_state
    print("maximum Accuracy is at random state :",rs)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
```

```python
svc = SVC()
maxf1_score(svc,df_x,df_y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_x, df_y,random_state = 43,test_size=0.20,stratify=df_y)
score=cross_val_score(svc,df_x,df_y,cv=5,scoring='accuracy')
print('accuracy Score=',score)
print('Mean Scores',score.mean())
print('Standard Deviation',score.std())

y_pred = cross_val_predict(svc,df_x,df_y,cv=5)

print('Y Prediction values')
print(y_pred)

conf_mat = confusion_matrix(df_y,y_pred)
conf_mat

accuracy_score(df_y,y_pred)
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_x, df_y,random_state = 45,test_size=0.20,stratify=df_y)
parameters = {'learning_rate':[0.1] ,  
              'min_samples_split':[5], 
              'min_samples_leaf': [50],
             'max_depth':[50],
             'max_features':['sqrt'],
             'subsample':[0.8]}  
  
gb_gs = GridSearchCV(GradientBoostingClassifier(), parameters) 
gb_gs.fit(x_train,y_train)
```

```python
from sklearn import ensemble
params = {'n_estimators': 500, 'max_depth': 4, 'min_samples_split': 2,
          'learning_rate': 0.01, 'loss': 'ls'}
gbr = ensemble.GradientBoostingRegressor(**params)

gbr.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
model_score = gbr.score(x_train,y_train)
print("Training_Score:",model_score)

print('R2 score: ',model_score)
y_pred = gbr.predict(x_test)

print("Mean squared error:", mean_squared_error(y_test, y_pred))
print('Test Variance score: %.2f' % r2_score(y_test, y_pred))
```
